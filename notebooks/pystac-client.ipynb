## Microsoft Planetry Computer

Connect to the Microsoft Planetry Computer STAC API

```python
catalog = Client.open("https://planetarycomputer.microsoft.com/api/stac/v1")
```

Search for items in the 3dep-seamless and nasadem collections by bounding box and time.

```python
time_range = "2020-12-01/2020-12-31"
bbox=[-78.89381702544192, 35.69930368900292, -78.5201720557924, 35.869395802052324]
my_search = catalog.search(
    max_items=10,
    collections=[
        '3dep-seamless',
        "nasadem",
        "3dep-lidar-dtm"
    ],
    bbox=bbox,
    datetime=time_range
)
print(f"{my_search.get_all_items()} items found")
items = my_search.get_all_items()
len(items)
```

Connect to the 3dep-seamless collection.

```python
collection = catalog.get_collection("3dep-seamless")
```

## USGS Lidar Catalog

Connect to the USGS lidar STAC catalog.

```python
usgs_catalog = Client.open("https://s3-us-west-2.amazonaws.com/usgs-lidar-stac/ept/catalog.json")
```

Get items in the usgs catalog.

```python
collection = usgs_catalog.get_items()
```

Search for items from North Carolina

```python
for item in collection:
    state = item.id.split("_")[0]
    if "NC" in state: 
        print(f"{state}: {item.id}")
```
